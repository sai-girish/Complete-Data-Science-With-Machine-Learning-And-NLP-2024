```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/content/Warehouse_and_Retail_Sales.csv')
```

```python
df
```

```python
df_beer=df[['YEAR','MONTH','ITEM TYPE','RETAIL SALES']]
```

```python
df_beer
```

```python
le = LabelEncoder()
df_beer['ITEM TYPE'] = le.fit_transform(df_beer['ITEM TYPE'])
```

```python
df_beer.isnull().sum()
```

```python
df_beer['RETAIL SALES'].mean()
```

```python
df_beer['RETAIL SALES']=df_beer['RETAIL SALES'].fillna(7.02)
```

```python
df_beer.isnull().sum()
```

```python
df_beer
```

```python
X = df_beer[['YEAR', 'MONTH', 'ITEM TYPE']]  # Input features
y = df_beer['RETAIL SALES']                  # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
future_data = pd.DataFrame({
    'YEAR': [2025],
    'MONTH': [5],
    'ITEM TYPE': le.transform(['BEER'])
})

future_sales = model.predict(future_data)
print("Predicted Sales for May 2025:", future_sales)
```

```python
future_data = pd.DataFrame({
    'YEAR': [2020],
    'MONTH': [5],
    'ITEM TYPE': le.transform(['BEER'])
})

future_sales = model.predict(future_data)
print("Predicted Sales for May 2020:", future_sales)
```
